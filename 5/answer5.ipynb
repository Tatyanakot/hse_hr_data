```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df_hr = pd.read_csv('WA.csv')

```

```python
# Отрисуем таблицу
pd.options.display.max_columns = 999
df_hr
```

```python
# Проверим датасет на наличие пустых значений
df_hr.info()
```

```python
# Проверим, является ли распределение сотрудников по уровню образованию нормальным
# H0: образование сотрудников распределено по нормальному закону
# H1: закон распределения уровня образования сотрудников не является нормальным

print (stats.shapiro(df_hr['Education']))
```

Раз значение p-value < 0.05 - H0 отклоняется. 
Следовательно, принимается H1: закон распределения уровня образования сотрудников в компании не является нормальным.

```python
# Построим диаграмму
plt.hist(df_hr['Education'])
```

Диаграмма также показывает, что распределение уровня образования сотрудников в компании не является нормальным.
